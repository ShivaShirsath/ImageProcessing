```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply contrast stretching
p2, p98 = np.percentile(img, (2, 98))
img_contrast_stretch = cv2.normalize(img, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)
img_contrast_stretch = cv2.convertScaleAbs(img_contrast_stretch)

# Apply intensity level slicing
lower_threshold = 100
upper_threshold = 200
img_intensity_slice = np.zeros_like(img)
img_intensity_slice[(img >= lower_threshold) & (img <= upper_threshold)] = 255
```

```python
# Display the results
cv2_imshow(img)
cv2_imshow(img_contrast_stretch)
cv2_imshow(img_intensity_slice)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
