```python
import cv2
from google.colab.patches import cv2_imshow

# Load input image
img = cv2.imread('image.jpg')
```

```python
# Define compression quality levels
jpeg_quality = 90
png_compression = 3
webp_quality = 90

# Compress using JPEG
jpeg_params = [int(cv2.IMWRITE_JPEG_QUALITY), jpeg_quality]
success, jpeg_img = cv2.imencode('.jpg', img, jpeg_params)
if success:
    with open('image_jpeg.jpg', 'wb') as f:
        f.write(jpeg_img)

# Compress using PNG
png_params = [int(cv2.IMWRITE_PNG_COMPRESSION), png_compression]
success, png_img = cv2.imencode('.png', img, png_params)
if success:
    with open('image_png.png', 'wb') as f:
        f.write(png_img)

# Compress using WebP
webp_params = [int(cv2.IMWRITE_WEBP_QUALITY), webp_quality]
success, webp_img = cv2.imencode('.webp', img, webp_params)
if success:
    with open('image_webp.webp', 'wb') as f:
        f.write(webp_img)
```

```python
# Display compressed images
cv2_imshow(cv2.imread('image_jpeg.jpg'))
cv2_imshow(cv2.imread('image_png.png'))
cv2_imshow(cv2.imread('image_webp.webp'))
```
